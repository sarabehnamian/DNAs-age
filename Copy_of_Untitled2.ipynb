```
from sklearn.neighbors import KNeighborsRegressor
```

```
from sklearn.model_selection import train_test_split
```

```
knn = KNeighborsRegressor()
```

```
country = df['Country']
```

```
import pandas as pd
```

```
dummies = pd.get_dummies(df['Country'], prefix='country')
```

```
df1 = pd.concat([dummies, df], axis=1)
```

```
df1
```

```
del df1['Country']
```

```
df1
```

```
X_train, X_test, y_train, y_test=train_test_split(df1, y, test_size=.25)
```

```
knn.fit(X_train, y_train)
```

```
from sklearn.metrics import mean_squared_error, r2_score
```

```
import math
```

```
math.sqrt(mean_squared_error(y_test, knn.predict(X_test)))
```

```
r2_score(y_test, knn.predict(X_test))
```
